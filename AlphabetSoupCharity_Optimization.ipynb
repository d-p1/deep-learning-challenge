# **Optimize the Model**

```python
# Import our dependencies
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import pandas as pd
import tensorflow as tf

#  Import and read the charity_data.csv.
import pandas as pd
application_df = pd.read_csv("https://static.bc-edx.com/data/dl-1-2/m21/lms/starter/charity_data.csv")
application_df.head()
```

```python
# Drop the non-beneficial ID columns, 'SPECIAL_CONSIDERATIONS'
application_df = application_df.drop(columns=['SPECIAL_CONSIDERATIONS'], axis=1)
application_df.head()
```

```python
#Creating more bins for rare occurrences in columns.
# Determine the number of unique values in each column.
application_df.nunique()
```

```python
# Look at APPLICATION_TYPE value counts for binning
app_counts = application_df['APPLICATION_TYPE'].value_counts()
app_counts
```

```python
#Increasing or decreasing the number of values for each bin.
```

```python
# Choose a cutoff value and create a list of application types to be replaced
# use the variable name `application_types_to_replace`
application_types_to_replace = list(app_counts[app_counts < 500].index)


# Replace in dataframe
for app in application_types_to_replace:
    application_df['APPLICATION_TYPE'] = application_df['APPLICATION_TYPE'].replace(app,"Other")

# Check to make sure binning was successful
application_df['APPLICATION_TYPE'].value_counts()
```

```python
# Look at CLASSIFICATION value counts for binning
class_counts_binning = application_df['CLASSIFICATION'].value_counts()

class_counts_binning.head()
```

```python
# You may find it helpful to look at CLASSIFICATION value counts >1
greaterThanClass = class_counts_binning.loc[class_counts_binning > 1]

print(greaterThanClass)
```

```python
# Choose a cutoff value and create a list of classifications to be replaced
# use the variable name `classifications_to_replace`
classifications_to_replace= list(class_counts_binning[class_counts_binning< 1000].index)


# Replace in dataframe
for cls in classifications_to_replace:
    application_df['CLASSIFICATION'] = application_df['CLASSIFICATION'].replace(cls,"Other")

# Check to make sure binning was successful
application_df['CLASSIFICATION'].value_counts()
```

```python

# Look at ORGANIZATION value counts for binning
org_counts = application_df['ORGANIZATION'].value_counts()
org_counts
```

```python
# Choose a cutoff value and create a list of organizations to be replaced
org_to_replace = list(org_counts[org_counts < 100].index)

# Replace in dataframe
for organization in org_to_replace:
    application_df['ORGANIZATION'] = application_df['ORGANIZATION'].replace(
        organization, "Other")

# Check to make sure binning was successful
application_df['ORGANIZATION'].value_counts()
```

```python
# Convert categorical data to numeric with `pd.get_dummies`
app_dummies_df = pd.get_dummies(application_df)
```

```python
# Split our preprocessed data into our features and target arrays
y = app_dummies_df['IS_SUCCESSFUL']
X = app_dummies_df.drop(columns = 'IS_SUCCESSFUL', axis = 1).values

# Split the preprocessed data into a training and testing dataset
rand_state = 1
X_train, X_test, y_train,y_test = train_test_split(X,y,random_state=rand_state)
```

```python
# Create a StandardScaler instances
scaler = StandardScaler()

# Fit the StandardScaler
X_scaler = scaler.fit(X_train)

# Scale the data
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
```

```python
# Define the model - deep neural net, i.e., the number of input features and hidden nodes for each layer.
#  YOUR CODE GOES HERE
input_ft_num = X_train.shape[1]
hidden_nodes1 =  10
hidden_nodes2 =  20
nn = tf.keras.models.Sequential()

# First hidden layer
nn.add(tf.keras.layers.Dense(units = hidden_nodes1,
                                   input_dim = input_ft_num,
                                   activation = 'tanh'))
# Second hidden layer
nn.add(tf.keras.layers.Dense(units=hidden_nodes2,
                             activation="relu"))

# Output layer
nn.add(tf.keras.layers.Dense(units=1,
                             activation="sigmoid"))


# Check the structure of the model
nn.summary()
```

```python
# Compile the model
compiled_model= nn.compile(loss="binary_crossentropy",
           optimizer="adam", metrics=["accuracy"])
```

```python
# Train the model
fit_model = nn.fit(X_train_scaled, y_train, epochs = 50)
```

```python
# Evaluate the model using the test data
model_loss, model_accuracy = nn.evaluate(X_test_scaled,y_test,verbose=2)
print(f"Loss: {model_loss}, Accuracy: {model_accuracy}")
```
